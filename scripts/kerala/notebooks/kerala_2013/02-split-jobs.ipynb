```python
from glob import glob
import os
import shutil
import pandas as pd
```

```python
N = 280
YEAR = 2013
```

```python
count = 0
files = []
for fn in glob('/home/t-9adar/gs-delhi/share/gcs/kerala/%d/*.pdf' % YEAR):
    count += 1
    files.append({'fn': fn})
count
```

```python
pd.options.display.max_colwidth = 100
```

```python
df = pd.DataFrame(files)
df
```

```python
import re
    
def extract_acno(c):
    m = re.match('.*?(\d+)_(\d+)_(\d+).*', c)
    if m:
        return int(m.group(2))
    return 0

def extract_partno(c):
    m = re.match('.*?(\d+)_(\d+)_(\d+).*', c)
    if m:
        return int(m.group(3))
    return 0


df['acno'] = df.fn.apply(lambda c: extract_acno(c))
df['partno'] = df.fn.apply(lambda c: extract_partno(c))
df
```

```python
import numpy as np

sdf = df.sort_values(['acno', 'partno'])
sdf
```

```python
for i, xdf in enumerate(np.array_split(sdf, N)):
    print(i)
    # FIXME: create dummy output folder
    path = '../../parse_unsearchable_rolls/data/part%d/out/kerala/%d/' % (i + 1, YEAR)
    if not os.path.exists(path):
        os.makedirs(path)        
    path = '../../parse_unsearchable_rolls/data/part%d/in/kerala/%d/' % (i + 1, YEAR)
    if not os.path.exists(path):
        os.makedirs(path)
    for idx, r in xdf.iterrows():
        #print(r.fn)
        try:
            shutil.copy(r.fn, path)
        except:
            pass
```
