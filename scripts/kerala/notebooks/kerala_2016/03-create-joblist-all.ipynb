```python
import os
from glob import glob
```

```python
N = 280
state = 'kerala'
```

```python
for i in range(1, N + 1):
    cmd = """python scripts/{0:s}/{0:s}_2016.py data/part{1:d}/
""".format(state, i)
    print(cmd)
    path = '../../parse_unsearchable_rolls/scripts/{0:s}/joblist_2016/{0:s}{1:d}.sh'.format(state, i)
    with open(path, 'w') as f:
        f.write(cmd)
    os.chmod(path, 0o755)
```

```python
for i in range(0, N):
    print(i)
    path = '../../parse_unsearchable_rolls/data/part%d/out/kerala/2016/' % (i + 1)
    if not os.path.exists(path):
        os.makedirs(path)
```
