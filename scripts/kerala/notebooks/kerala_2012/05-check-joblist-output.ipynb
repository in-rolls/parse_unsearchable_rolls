```python
import re
from glob import glob
import pandas as pd
```

```python
data = []
for fn in sorted(glob('../../parse_in_rolls/scripts/kerala/joblist-output/KeralaJobArray.o311542-*')):
    #print(fn)
    job_no = int(fn.split('-')[-1])
    with open(fn) as f:
        txt = f.read()
        for m in re.finditer('(.*?\.pdf) (\d+/\d+/\d+ \d+:\d+:\d+).*?\n.*?\n.*?\n.*?\.pdf (\d+/\d+/\d+ \d+:\d+:\d+).*?\n.*?: finish in (\d+)s', txt):
            file = m.group(1)
            start = m.group(2)
            finish = m.group(3)
            duration = int(m.group(4))
            data.append({'job_no': job_no, 'file': file, 'duration': duration, 'start': start, 'finish': finish})
```

```python
df = pd.DataFrame(data)
df
```

```python
df['start']  = pd.to_datetime(df.start)
df['finish'] = pd.to_datetime(df.finish)
```
