```python
from glob import glob
import pandas as pd
```

```python
files = []
for i in range(1, 281):
    files.extend([{'fn': fn} for fn in glob('/share/svasudevan2lab/gs-delhi/parse_unsearchable_rolls/data/part%d/out/kerala/2011/*.csv' % i)])
df = pd.DataFrame(files)
df
```

```python
import re
    
def extract_acno(c):
    m = re.match('.*?(\d+)_(\d+)_(\d+).*', c)
    if m:
        return int(m.group(2))
    return 0

def extract_partno(c):
    m = re.match('.*?(\d+)_(\d+)_(\d+).*', c)
    if m:
        return int(m.group(3))
    return 0


df['acno'] = df.fn.apply(lambda c: extract_acno(c))
df['partno'] = df.fn.apply(lambda c: extract_partno(c))
df
```

```python
sdf = df.sort_values(['acno', 'partno']).reset_index(drop=True)
sdf
```

```python
import os

for i, r in sdf.iterrows():
    s = os.path.getsize(r.fn)
    if s == 0:
        print(r.fn)
```

```python
for i, r in sdf.iterrows():
    odf = pd.read_csv(r.fn)
    if i == 0:
        odf.to_csv('/share/svasudevan2lab/gs-delhi/parse_unsearchable_rolls/data/out/kerala_2011.csv.gz', index=False, compression='gzip')
    else:
        odf.to_csv('/share/svasudevan2lab/gs-delhi/parse_unsearchable_rolls/data/out/kerala_2011.csv.gz', mode='a', header=False, index=False, compression='gzip')
    if i % 1000 == 0:
        print(i)
print(i)
```
