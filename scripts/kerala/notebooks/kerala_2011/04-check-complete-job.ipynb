```python
from subprocess import Popen, PIPE
import shlex

def execute(cmd):
    p = Popen(shlex.split(cmd), stdout=PIPE, stderr=PIPE, close_fds=True)
    out, err = p.communicate()
    return out.decode('utf-8'), err.decode('utf-8')
```
